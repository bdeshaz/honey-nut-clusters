```python
class Scaled:

    def __init__(self):
        pass
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
#         return X.truediv(X.cups, axis='index')
        cups = X[:,-1]
        cups = cups.reshape(77,1)
        return X / cups
```

```python
estimator = Pipeline([("imputer", preprocessing.Imputer()),
                      ("scaled", Scaled()),
                      ("normalize", Normalizer()),
                      ('kmeans', KMeans(4))])
```

```python
estimator.fit_transform(data)
```

```python
estimator.predict(data)
```
